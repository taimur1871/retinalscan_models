<h2>Edge Detection Techniques</h2>
Trying various edge detectors for cutter wear measure

```
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```
img = cv.imread('Bld969.jpg',0)
edges = cv.Canny(img,100,150, L2gradient=True)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```
import numpy as np

import tensorflow as tf
from tensorflow import image
from PIL import Image
```

```
# convert im to tensor
imgs = Image.open('Bld969.jpg')
im_tensor = np.asarray(imgs)
im_tensor = np.expand_dims(im_tensor, axis=0)
im_tensor = tf.convert_to_tensor(im_tensor, dtype=float)
im_tensor.get_shape()
```

```
sobel_tensor = image.sobel_edges(im_tensor)
```

```
sobel_np = sobel_tensor[0, :, :, :, 1].numpy()
```
